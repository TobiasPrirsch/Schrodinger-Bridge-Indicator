```python
episode = 800
tra = np.load(f'train_itf{episode}.npy')
plt.figure(figsize=(20,30))
plt.subplot(6, 2, 1)
# plt.figure(figsize=(4,3))
plt.scatter(tra[:, 0, 0], tra[:, 0, 1], label=f'1 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.legend()
# plt.show()

for i in range(5):
    num = (i+1)*20-1
    plt.subplot(6, 2, i*2+3)
#     plt.figure(figsize=(4,3))
    
    plt.scatter(tra[:, num, 0], tra[:, num, 1], label=f'{(i+1)*20} step')
    plt.xlim(-100,200)
#     plt.xlim(-50,50)
#     plt.ylim(-50,50)
    plt.ylim(-120,120)
    plt.legend()
#     plt.show()
    
tra = np.load(f'train_itb{episode}.npy')
plt.subplot(6, 2, 2)
# plt.figure(figsize=(4,3))
plt.scatter(tra[:, 0, 0], tra[:, 0, 1], label=f'1 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.legend()
# plt.show()

for i in range(5):
    num = (i+1)*20-1
    plt.subplot(6, 2, i*2+4)
#     plt.figure(figsize=(4,3))
    plt.scatter(tra[:, num, 0], tra[:, num, 1], label=f'{(i+1)*20} step')
    plt.xlim(-100,200)
#     plt.xlim(-50,50)
    plt.ylim(-120,120)
    plt.legend()
plt.show()
```
