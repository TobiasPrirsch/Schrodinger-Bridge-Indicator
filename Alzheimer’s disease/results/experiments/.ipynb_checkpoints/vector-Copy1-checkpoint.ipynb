```python
step = 20
grad = zf_right[800,:,step,:]
grad.shape
```

```python
gradient = grad[:50,]
length = (gradient[:,0]**2 + gradient[:,1]**2)**0.5 / 50
```

```python
direction_vector = gradient / 20
point = tra[:50,step,:]
end_point = point + direction_vector 
```

```python
# 绘制点和矢量  
plt.figure(figsize=(10, 10))  
plt.scatter(point[:,0], point[:,1], s=200, color='pink', zorder=1)  # 绘制点  
plt.quiver(point[:,0], point[:,1], direction_vector[:,0], direction_vector[:,1], angles='xy', scale_units='xy', color='gray',alpha=0.8,
           headwidth=5, headlength=6, linewidths=0.5, scale=0.7)  # 绘制矢量  
plt.xlim(-0,50)
plt.ylim(-50,50)
```
