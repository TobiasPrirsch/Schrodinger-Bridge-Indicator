```python
plt.rcParams['font.family'] = 'Arial'
```

```python
plt.figure(figsize=(8,6))
loss_left = np.load('loss_left.npy')
plt.plot(loss_left[20:], label='Loss')
plt.legend(fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
loss_right = np.load('loss_right.npy')
plt.plot(loss_right[20:], label='Loss')
plt.legend(fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
loss_total = loss_left + loss_right
plt.plot(loss_total[20:], label='Loss')
plt.legend(fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# plt.savefig('ad_loss.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
episode = 1000
fontsize=15
fontsize1=15
tra_fleft = np.load(f'left/train_itf{episode}.npy')
tra_fright = np.load(f'right/train_itf{episode}.npy')
plt.figure(figsize=(30,10))

plt.subplot(2, 5, 1)
plt.scatter(tra_fleft[:, 0, 0], tra_fleft[:, 0, 1], s=5, c='cornflowerblue', label=f'1 step')
plt.scatter(tra_fright[:, 0, 0], tra_fright[:, 0, 1], s=5, c='orange', label=f'1 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')

plt.subplot(2, 5, 2)
plt.scatter(tra_fleft[:, 25, 0], tra_fleft[:, 25, 1], s=5, c='cornflowerblue', label=f'25 step')
plt.scatter(tra_fright[:, 25, 0], tra_fright[:, 25, 1], s=5, c='orange', label=f'25 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')

plt.subplot(2, 5, 3)
plt.scatter(tra_fleft[:, 50, 0], tra_fleft[:, 50, 1], s=5, c='cornflowerblue', label=f'50 step')
plt.scatter(tra_fright[:, 50, 0], tra_fright[:, 50, 1], s=5, c='orange', label=f'50 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper right')

plt.subplot(2, 5, 4)
plt.scatter(tra_fleft[:, 75, 0], tra_fleft[:, 75, 1], s=5, c='cornflowerblue', label=f'75 step')
plt.scatter(tra_fright[:, 75, 0], tra_fright[:, 75, 1], s=5, c='orange', label=f'75 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper right')

plt.subplot(2, 5, 5)
plt.scatter(tra_fleft[:, -1, 0], tra_fleft[:, -1, 1], s=5, c='cornflowerblue', label=f'100 step')
plt.scatter(tra_fright[:, -1, 0], tra_fright[:, -1, 1], s=5, c='orange', label=f'100 step')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')


tra_bleft = np.load(f'left/train_itb{episode}.npy')
tra_bright = np.load(f'right/train_itb{episode}.npy')

plt.subplot(2, 5, 6)
plt.scatter(tra_bleft[:, 0, 0], tra_bleft[:, 0, 1], s=5, c='salmon', label=f'1 step')
plt.scatter(tra_bright[:, 0, 0], tra_bright[:, 0, 1], s=5, c='salmon', )
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')

plt.subplot(2, 5, 7)
plt.scatter(tra_bleft[:, 25, 0], tra_bleft[:, 25, 1], s=5, c='salmon', label=f'25 step')
plt.scatter(tra_bright[:, 25, 0], tra_bright[:, 25, 1], s=5, c='salmon')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')

plt.subplot(2, 5, 8)
plt.scatter(tra_bleft[:,50, 0], tra_bleft[:, 50, 1], s=5, c='salmon', label=f'50 step')
plt.scatter(tra_bright[:, 50, 0], tra_bright[:, 50, 1], s=5, c='salmon')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper right')

plt.subplot(2, 5, 9)
plt.scatter(tra_bleft[:, 75, 0], tra_bleft[:, 75, 1], s=5, c='salmon', label=f'75 step')
plt.scatter(tra_bright[:, 75, 0], tra_bright[:, 75, 1], s=5, c='salmon')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper right')


plt.subplot(2, 5, 10)
plt.scatter(tra_bleft[:, -1, 0], tra_bleft[:, -1, 1], s=5, c='salmon', label=f'100 step')
plt.scatter(tra_bright[:, -1, 0], tra_bright[:, -1, 1], s=5, c='salmon')
plt.xlim(-100,200)
plt.ylim(-120,120)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize1, loc='upper left')
# plt.savefig('ad_joint.png', bbox_inches='tight', dpi=150)
plt.show()
```
