```python
zf_right = np.load('zf_right.npy')
zf_left = np.load('zf_left.npy')
zf_right.shape, zf_left.shape
```

```python
zf = np.concatenate([zf_left,zf_right],axis=1)
zf.shape
```

```python
zf = np.mean(zf[:,:,:,0]**2 + zf[:,:,:,1]**2, axis=(1))
zf_left = np.mean(zf_left[:,:,:,0]**2 + zf_left[:,:,:,1]**2, axis=(1))
zf_right = np.mean(zf_right[:,:,:,0]**2 + zf_right[:,:,:,1]**2, axis=(1))
```

```python
ee = 1999
inte = np.cumsum((zf[ee, :])*0.01)
temp = []
for i in range(len(inte)):
    temp.append(inte[i]/(0.01*(i+1)))
```

```python
plt.rcParams['font.family'] = 'Arial'
```

```python
plt.figure(figsize=(8,6))
plt.scatter(range(100), temp, c='r', label='Indicator', s=10)
plt.vlines(x=24, ymin=temp[0], ymax=temp[24], label='25 step')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=18)
# plt.savefig('ad_indicator.png',bbox_inches='tight',dpi=150)
plt.show()
```
