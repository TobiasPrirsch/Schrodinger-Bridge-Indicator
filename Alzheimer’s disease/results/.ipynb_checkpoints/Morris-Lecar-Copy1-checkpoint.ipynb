```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
C = 20
phi = 0.23
I = 37
V_ca = 120
V_k = -84
V_l = -60
g_ca = 4
g_k = 8
g_l = 2
V1 = -1.2
V2 = 18
V3 = 12
V4 = 17.4
```

```python
def m_inf(v):

    temp = (v - V1) / V2
    
    return 0.5 * (1 + np.tanh(temp))

def w_inf(v):

    temp = (v - V3) / V4
    
    return 0.5 * (1 + np.tanh(temp))

def tau_w(v):
    
    temp = (v - V3) / (2 * V4)
    return np.cosh(temp) ** -1
```

```python
# (-4*0.5*(1+tanh((x+1.2)/18))*(x-120)-8*y*(x+84)-2*(x+60)+a)/20
# 0.04*((0.5*(1+tanh(x-2)/30))-y)*(cosh((x-2)/60))
```

```python
def fun_vt(data, dt):
    
    v = data[:, 0]
    w = data[:, 1]
    drift = (- g_ca * m_inf(v) * (v - V_ca) - g_k * w * (v - V_k) - g_l * (v - V_l) + I) / C
#     v_next = v +  drift * dt + sigma * dw
    v_next = v +  drift * dt
    
    return v_next

def fun_wt(data, dt):
    
    v = data[:, 0]
    w = data[:, 1]
    drift = phi * (w_inf(v) - w) / tau_w(v)
    w_next = w +  drift * dt
    
    return w_next
```

```python
T = 10
interval = 10000
dt = 0.1
dw = np.sqrt(dt) * np.random.normal(size=(1,))
sigma = 0.01
```

```python
x0 = np.array([[-40, 0.]])
tra_v = []
tra_w = []
tra_v.append(x0[:, 0])
tra_w.append(x0[:, 1])
# v = x0[:, 0]
# w = x0[:, 1]
data = x0
for i in range(interval):
    v = fun_vt(data, dt)
    w = fun_wt(data, dt)
    data = np.concatenate([v.reshape(-1,1), w.reshape(-1,1)], axis=1)
    tra_v.append(v)
    tra_w.append(w)
plt.scatter(np.array(tra_v), np.array(tra_w),s=1)
plt.xlim(-40,-30)
plt.ylim(-0.3,0.6)
```
