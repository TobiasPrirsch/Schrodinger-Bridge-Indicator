```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
```

```python
mild = np.load('./experiments/mildx.npy')
```

```python
# 生成示例二维散点数据
# np.random.seed(0)
# x = np.random.normal(size=1000)
# y = np.random.normal(size=1000)
x = mild[:, 0]
y = mild[:, 1]

# 将散点数据合并成一个二维数组
data = np.vstack([x, y])

# 使用gaussian_kde函数进行核密度估计
kde = gaussian_kde(data)

# 生成二维网格，用于绘制密度估计结果
x_grid, y_grid = np.meshgrid(np.linspace(x.min(), x.max(), 100),
                              np.linspace(y.min(), y.max(), 100))
grid_coords = np.vstack([x_grid.ravel(), y_grid.ravel()])

# 计算每个网格点的密度值
density_values = kde(grid_coords).reshape(x_grid.shape)
```

```python
# 绘制密度估计结果
plt.figure(figsize=(8, 6))
plt.imshow(density_values, origin='lower', aspect='auto',
           extent=[x.min(), x.max(), y.min(), y.max()], cmap='viridis')
plt.colorbar(label='Density')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Kernel Density Estimation')
plt.show()
```
