```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steady = np.load('data_steady.npy')
cycle = np.load('data_cycle.npy')
steady.shape, cycle.shape
```

```python
plt.rcParams['font.family'] = 'Arial'
```

```python
from matplotlib.lines import Line2D
```

```python
tra01 = np.load('./3 100 0.1/train_itf1000.npy')
tra03 = np.load('./3 100 0.3/train_itf1000.npy')
tra05 = np.load('./3 100 0.5/train_itf1000.npy')
tra1 = np.load('./3 100 1/train_itf1000.npy')
tra2 = np.load('./3 100 2/train_itf1000.npy')
tra3 = np.load('./3 100 3/train_itf1000.npy')
tra4 = np.load('./3 100 4/train_itf1000.npy')
tra5 = np.load('./3 100 5/train_itf1000.npy')
```

```python
plt.figure(figsize=(12,8))
plt.scatter(cycle[:,0], cycle[:,1], s=5)
plt.scatter(tra01[:, -1, 0][:], tra01[:, -1, 1][:], s=5, zorder=5, color='darkorange',label='g = 0.1')
plt.scatter(tra05[:, -1, 0][:], tra05[:, -1, 1][:], s=5, zorder=3, color='g',label='g = 0.5')
plt.scatter(tra1[:, -1, 0][:], tra1[:, -1, 1][:], s=5, zorder=2, color='r',label='g = 1.0')
plt.scatter(tra3[:, -1, 0], tra3[:, -1, 1], s=5, zorder=0,  color='purple',label='g = 3.0')
plt.scatter(tra4[:, -1, 0], tra4[:, -1, 1], s=5, zorder=0,  color='grey',label='g = 4.0')
plt.scatter(tra5[:, -1, 0], tra5[:, -1, 1], s=5, color='brown')

line01 = Line2D([0], [0], color='darkorange', label='g = 0.1')
line05 = Line2D([0], [0], color='g', label='g = 0.5')
line1 = Line2D([0], [0], color='r', label='g = 1.0')
line3 = Line2D([0], [0], color='purple', label='g = 3.0')
line4 = Line2D([0], [0], color='grey', label='g = 4.0')
line5 = Line2D([0], [0], color='brown', label='g = 5.0')

plt.legend(fontsize=20, handles=[line01, line05, line1, line3, line4, line5])

plt.xlim(-60,60)
plt.ylim(0, 0.6)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# plt.savefig('shift_noise2.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
size = 50
size1= 300
fontsize=35
tra = np.load('./3 100 0.1/train_itf1000.npy')
plt.figure(figsize=(80,10))

plt.subplot(1, 5, 1)
plt.scatter(cycle[:,0], cycle[:,1], s=size)
plt.scatter(tra[:, 0, 0], tra[:, 0, 1], s=size1, c='r', label=f'1 step')
plt.xlim(-60,40)
plt.ylim(0, 0.6)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize, loc='upper left')

plt.subplot(1, 5, 2)
plt.scatter(cycle[:,0], cycle[:,1], s=size)
plt.scatter(tra[:, 24, 0], tra[:, 24, 1], s=size1, c='r',  label=f'25 step')
plt.xlim(-60,40)
plt.ylim(0, 0.6)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize, loc='upper left')

plt.subplot(1, 5, 3)
plt.scatter(cycle[:,0], cycle[:,1], s=size)
plt.scatter(tra[:, 49, 0], tra[:, 49, 1], s=size1, c='r',  label=f'50 step')
plt.xlim(-60,40)
plt.ylim(0, 0.6)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize, loc='upper left')

plt.subplot(1, 5, 4)
plt.scatter(cycle[:,0], cycle[:,1], s=size)
plt.scatter(tra[:, 74, 0], tra[:, 74, 1], s=size1, c='r',  label=f'75 step')
plt.xlim(-60,40)
plt.ylim(0, 0.6)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize, loc='upper left')

plt.subplot(1, 5, 5)
plt.scatter(cycle[:,0], cycle[:,1], s=size)
plt.scatter(tra[:, -1, 0], tra[:, -1, 1], s=size1,c='r',  label=f'100 step')
plt.xlim(-60,40)
plt.ylim(0, 0.6)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize, loc='upper left')
# plt.savefig('3_100_01.png', bbox_inches='tight', dpi=150)
plt.show()
```
