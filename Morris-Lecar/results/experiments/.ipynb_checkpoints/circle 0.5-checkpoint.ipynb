```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cycle = np.load('data_cycle.npy')
tra10 = np.load('./10 100 0.5/train_itf1600.npy')
tra30 = np.load('./30 300 1/train_itf1600.npy')
tra40 = np.load('./40 200 0.6 1 1-2/train_itf800.npy')
```

```python
from matplotlib.lines import Line2D
```

```python
plt.figure(figsize=(12,8))
plt.scatter(cycle[:,0], cycle[:,1], s=5)
scatter10 = plt.scatter(tra10[:, -1, 0], tra10[:, -1, 1], c='darkorange', s=5)
scatter30 = plt.scatter(tra30[:, -1, 0], tra30[:, -1, 1], c='g', s=5, label='T=30')
scatter40 = plt.scatter(tra40[:, -1, 0], tra40[:, -1, 1], c='r', s=5, label='T=40')
line10 = Line2D([0], [0], color='darkorange', label='T=10')
line30 = Line2D([0], [0], color='g', label='T=30')
line40 = Line2D([0], [0], color='r', label='T=40')

plt.xlim(-60,60)
plt.ylim(0, 0.6)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=20, handles=[line10, line30, line40])

# plt.savefig('shift_T.png', bbox_inches='tight', dpi=150)
plt.show()
```
