```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
c = 20
V_c = 120
V_k = -84
V_L = -60
g_c = 4.4 
g_k = 8
g_l = 2
v_1 = -1.2
v_2 = 18
v_3 = 2
v_4 = 30
phi = 0.04
I = 92
```

```python
data = np.load('data_steady.npy')
data = torch.tensor(data)
data.shape
```

```python
mu = torch.mean(data, axis=0).to(torch.float32)
sigma = torch.tensor([1e-2, 1e-4]) * torch.eye(2)
gauss_pdf = torch.distributions.MultivariateNormal(mu, sigma)
print(mu, sigma)
```

```python
T = 100
interval = 10000
delta_t = T / interval
g = 1
```

```python
x0 = gauss_pdf.sample([20])
v = x0[:, 0]
w = x0[:, 1]
v, w
```

```python
# log_v = []
# log_w = []

def mm(v):
    m = 0.5 * (1 + np.tanh((v - v_1)/v_2))
    return m

def www(v):
    ww = 0.5 * (1 + np.tanh((v - v_3)/v_4))
    return ww

def tau(v):
    ta = 1 / (np.cosh((v - v_3)/ 2 /v_4))
    return ta

for i in range(interval):
    v = 1 / c * (- g_c * mm(v) * (v - V_c) - g_k * w * (v - V_k) - g_l * (v - V_L) + I) * delta_t + v + np.sqrt(delta_t) * g * np.random.normal(loc=0, scale=1)
    w = phi * (www(v) - w) / tau(v) * delta_t + w 
#     log_v.append(v)
#     log_w.append(w)
v, w
```

```python
plt.scatter(v, w)
plt.xlim(-70,40)
plt.ylim(0,0.6)
plt.show()
```
